# Popularity Based Filtering

```python
ratings_name = ratings.merge(books ,on='ISBN') # merge ratings and books data set 
ratings_name
```

```python
# to remove multiple rating 
no_of_rating = ratings_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
no_of_rating.rename(columns={'Book-Rating':'num_of_rating'},inplace=True)
no_of_rating
```

```python
# Average rating of each book 
average_rating = ratings_name.groupby('Book-Title').mean()['Book-Rating'].reset_index()
average_rating.rename(columns={'Book-Rating':'avg_rating'},inplace=True)
average_rating
```

```python
# merge average rating and no_of_rating df  
popular_books = no_of_rating.merge(average_rating ,on='Book-Title')
popular_books
```

```python
# show all the books who has number of rating more then 500 
# here we get books who most users rated however the average rating is bad 
popular_books[popular_books['num_of_rating']>=500].sort_values('num_of_rating',ascending=False)
```

## Filter data by popularity 

```python
#filter data based on average rating & most number of ratings to get the most piopular books
popular_books_df=popular_books[popular_books['num_of_rating']>=500].sort_values('avg_rating',ascending=False)
popular_books_df
```

```python
# show popularity based filtering by Book-title, book-author and year of publication 
filtered_books = popular_books_df.merge(books ,on='Book-Title').drop_duplicates('Book-Title')[['ISBN','Book-Title','Book-Author','Year-Of-Publication', 'Image-URL-L']]
print(filtered_books.shape)
filtered_books.reset_index()
```

# Colloborative Filtering

```python
ratings_name.head()
```

```python
# filter users who gave more reviews to books 
a = ratings_name.groupby('User-ID').count()['Book-Title']>300 # 
geninue_users = a[a].index
geniune_user = ratings_name[ratings_name['User-ID'].isin(geninue_users)]
geniune_user
```

```python
# filter books who got more then 50 ratings 
b = geniune_user.groupby('Book-Title').count()['Book-Rating']>=50
famous_books = b[b].index
final_df = geniune_user[geniune_user['Book-Title'].isin(famous_books)]
final_df
```

```python
final_df.drop_duplicates()
```

```python
# prepare data to save to external file
newdf = final_df.drop(["User-ID","Book-Rating"], axis='columns')
newdf = newdf.drop_duplicates()
newdf.count()
```

```python
# reference: https://www.kaggle.com/code/jagdishchavan/book-recommender-system
pivot_table = final_df.pivot_table(index='Book-Title',columns='User-ID')
pivot_table.fillna(0,inplace=True)
pivot_table.head(500)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# caluclate eucledian distance for given data
cos_simscore = cosine_similarity(pivot_table)
print(cos_simscore.shape)
cos_simscore
```

```python
# function to recomend books 
def recommend(books_name):
    index = np.where(pivot_table.index==books_name)[0][0]    
    similar_books = sorted(list(enumerate(cos_simscore[index])),key=lambda x:x[1],reverse=True)[1:6]
    
    for i in similar_books:
        print(pivot_table.index[i[0]])
```

```python
recommend('The Fellowship of the Ring (The Lord of the Rings, Part 1)')
```

```python
def Book_info(title): # access book info
    book = final_df[final_df["Book-Title"] == str(title)]
    book = book.reset_index() 
    return book.loc[0]
```
